```python
dataf = dataf[(dataf['NSP'] >= 0) & (dataf['NSP'] <= 3)]
dataf['NSP'] = dataf['NSP'].round()
dataf.shape  
dataf['NSP'].value_counts()
```

### Handle missing values appropriately 

```python
# Check for missing values in each column of the DataFrame
dataf.isnull().sum()
```

```python
# We initialize SimpleImputer with the strategy='mean'.
# The fit_transform method is used to compute the mean of each column and replace missing values.
#Create a SimpleImputer with the mean strategy.
# For simplicity, we'll use mean imputation for numerical columns
from sklearn.impute import SimpleImputer
imputer =  SimpleImputer(strategy='mean')
dataf= pd.DataFrame(imputer.fit_transform(dataf),columns=dataf.columns)
```

```python
dataf
```

```python
# now check the transformed data set have null values or not
dataf.isnull().sum()
```

```python
# Print the current data types of each column in the DataFrame
print('current data types:')
print(dataf.dtypes)
```

```python
# Convert all columns to numeric data types, setting non-convertible values to NaN
dataf = dataf.apply(pd.to_numeric, errors='coerce')
dataf
```

### Detect and treat outliers

```python
# Plot a boxplot to visualize data before handling outliers
plt.figure(figsize=(12, 9))  # Set the figure size
dataf.boxplot()  # Create the boxplot for the DataFrame
plt.title('Boxplot Before Handling Outliers')  # Add a title
plt.show()  # Display the plot
```

```python
def treat_outliers(dataf, column):
    # Calculate the first quartile (25th percentile) of the column
    q1 = dataf[column].quantile(0.25)
    
    # Calculate the third quartile (75th percentile) of the column
    q3 = dataf[column].quantile(0.75)
    
    # Compute the Interquartile Range (IQR), which is the range between Q1 and Q3
    IQR = q3 - q1
    
    # Calculate the upper bound for outliers (Q3 + 1.5 * IQR)
    upper_extreme = q3 + 1.5 * IQR
    
    # Calculate the lower bound for outliers (Q1 - 1.5 * IQR)
    lower_extreme = q1 - 1.5 * IQR
    
    # Replace any values in the column that are less than the lower bound with the lower bound
    dataf[column] = np.where(dataf[column] < lower_extreme, lower_extreme, dataf[column])
    
    # Replace any values in the column that are greater than the upper bound with the upper bound
    dataf[column] = np.where(dataf[column] > upper_extreme, upper_extreme, dataf[column])
    
    # Return the modified DataFrame with outliers treated
    return dataf
```

```python
# Detect and treat outliers for each column
for column in dataf.columns:
    dataf = treat_outliers(dataf, column)  
dataf
```

```python
# Set the size of the figure (width=10, height=5)
plt.figure(figsize=(10, 5))

# Create a boxplot of the DataFrame columns to visualize the data after treating outliers
dataf.boxplot()

# Add a title to the plot to indicate it shows the boxplot after treating outliers
plt.title('Boxplot After Treating Outliers')

# Display the plot on the screen
plt.show()
```
